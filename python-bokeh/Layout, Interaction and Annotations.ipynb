# Layout, Interaction and Annotations
how to combine mutiple Bokeh plots into different kinds of layouts on a page, how to easily link different plots together in various ways, and how to add annotations such as legends and hover tooltips.
## Rows and Columns of plots

```python
from bokeh.layouts import row, column
from bokeh.sampledata.iris import flowers
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()

p1 = figure(plot_width = 400, plot_height=300)
p2 = figure(plot_width = 400, plot_height=300)
p3 = figure(plot_width = 400, plot_height=300)

p1.circle(flowers.sepal_length, flowers.sepal_width, color="black", alpha=0.5)
p2.circle(flowers.petal_length, flowers.petal_width, color="red", alpha=0.5)
p3.circle(flowers.sepal_length, flowers.petal_length, color="purple", alpha=0.5)

layout = row([p1, p2, p3])
show(layout)
```

```python
layout_2 = column([p1, p2, p3])
show(layout_2)
```

```python
layout_3 = row(column([p1,p2]), p3)
show(layout_3)
```

```python
from bokeh.layouts import row, column
from bokeh.models import ColumnDataSource
import pandas as pd
df = pd.read_csv("data/literacy_birth_rate.csv")
source = ColumnDataSource(df)
p1 = figure(x_axis_label="fertility", y_axis_label="female literacy", plot_height=300, plot_width=500)
p1.circle(x="fertility", y="female literacy", source=source)
p2 = figure(x_axis_label="popluation", y_axis_label="female literacy", plot_height=300, plot_width=500)
p2.circle(x="population", y="female literacy", source = source)
layout = row([p1, p2])
show(layout)
```
